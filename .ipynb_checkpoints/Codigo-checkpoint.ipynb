```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.gridspec as gridspec
import numpy as np

# Importa base de dados
df_titanic = pd.read_csv('Import/titanic.csv')

# Agrupa a tabela por classe e seleciona a coluna de survived para somar o total de passageiros de cada clase e fazer a média. Tendo assim  a taxa de sobrevivencia
# Taxa de Sobrevivência (%) = (Número de Sobreviventes do Grupo / Total de Passageiros no Grupo) * 100
resumo = df_titanic.groupby('Pclass')['Survived'].agg(['count', 'sum', 'mean']).reset_index()
# Renomeia Colunas
resumo.columns = ['Classe', 'Total de Passageiros', 'Sobreviventes', 'Taxa de Sobrevivência (%)']
#transforma a taxa de sobrevivencia em porcentagem
resumo['Taxa de Sobrevivência (%)'] = (resumo['Taxa de Sobrevivência (%)'] * 100).round(1)


#Preparando variaveis
classes = resumo['Classe'].astype(str) #formata campo como string
total = resumo['Total de Passageiros']
sobreviventes = resumo['Sobreviventes']
x = np.arange(len(classes)) #cria array com as possições de cada barra para 'Desenhar' no grafico depois
largura = 0.35 #largura das barras


fig, ax = plt.subplots()
#define qual informação vai em cada barra, bem como tamanho, legenda e cores
b1 = ax.bar(x - largura/2, total, width=largura, label='Total de Passageiros', color='skyblue')
b2 = ax.bar(x + largura/2, sobreviventes, width=largura, label='Sobreviventes', color='lightgreen')

ax.set_title('Total de Passageiros e Sobreviventes por Classe') #titulo do grafico
ax.set_xlabel('Classe') # Rotulo do eixo X
ax.set_ylabel('Quantidade') # Rotulo do eixo Y
ax.set_xticks(x) #Onde os rotulos do eixo X vao aparecer. Array criado anteriormente
ax.set_xticklabels(classes) # Quais textos vao aparecer nas possições do eixo X
ax.legend() # Legenda de cor do grafico


for barra in b1:
    altura = barra.get_height() #valor que representa a altura da barra
    ax.text(
            barra.get_x() + barra.get_width() / 2, # posição X - centraliza o texto na barra 
            altura + 3,  # posição Y - um pouco acima da barra
            str(int(altura)), #Transforma em numero inteiro
            ha='center', #Alinha na horizontal
            fontweight='bold' #texto em negrito
    )

for barra in b2:
    altura = barra.get_height()
    ax.text(
            barra.get_x() + barra.get_width() / 2,
            altura + 3, str(int(altura)), 
            ha='center', 
            fontweight='bold'
    )


total_geral = resumo['Total de Passageiros'].sum() #usado no final do grafico para mostrar o numero total de passageiros 

plt.tight_layout()
plt.figtext(0.5, # posição horizontal (50% do gráfico, ou seja, centralizado)
            -0.05,# posição vertical (um pouco abaixo do gráfico) 
            f'Total de passageiros a bordo: {total_geral}', # texto 
            ha='center', fontsize=10, fontweight='bold')

plt.savefig('Export/otal_passageiros_e_sobreviventes.png')
plt.show()




```
